#### 2.2. Video Example<a id='video_example_det2'></a>

[Sections](#sections_det2)

NOTE: For simplicity, the result will only be shown in an image format with only Panoptics Segmentation example.

```python
# create Detectron2 object for video use case
Detectron2 = Detectron2(model_type = "PS")
```

```python
# define params

folder_path = '/path_to_folder'

video = '/shibuya.mp4'
frame_num = '/frame1.jpg'

video_path = folder_path + video
frame_path = folder_path + frame_num

num_frames_extraction = 3

print('Read video on: ', video_path)
print('Save frames on: ', frame_path)
print('Expect saved frames: ', num_frames_extraction)
```

```python
# define function for video to frame extraction
def video_to_frame(video_path, num_frames, save_frame_folder_path):
    import cv2
    vidcap = cv2.VideoCapture(video_path)
    success,image = vidcap.read()
    count = 0
    
    while success and count < num_frames:
        cv2.imwrite(save_frame_folder_path + "/frame%d.jpg" % count, image)     # save frame as JPEG file      
        success,image = vidcap.read()
        print('Read a new frame: ', success)
        count += 1
    print('\nDone reading!')
```

```python
# frames extraction
video_to_frame(video_path, num_frames_extraction, folder_path)
```

##### A frame sample Demo

```python
# start segmentation
Detectron2.onImage(frame_path)
```

**Discussion & Closing:**

One of the points that is worth mentioning is Transfer Learning (TL), which is widely used in nearly all CV problems. TL is the idea of leveraging a pre-trained model (eps. weights) for a new problem without training the model from scatrch. This idea is particularly useful when solving complex problems for many reasons. Apart from overcoming the issue of insufficient dataset, TL also helps speeding up the deployment of any real-world applications without forgoing too much cost. For these reasons, many new domains with little knowledge and limited resources (e.g. skilled domain experts, a massive supporting dataset, powerful technologies) could now make use of Deep Learning models. For instance, this example has leveraged on the pre-trained models for different uses, thanks to the people who make the tools available for everyone.
